The following script creates a simple line chart showing how revenue changed over four months. It demonstrates all three layers of Matplotlib plotting: (1) Layer 1 (Data): Two lists – months for X-axis, revenue for Y-axis (2) Layer 2 (Options): Styling settings passed inside plt.plot() – color, line style, markers, and label (3) Layer 3 (Chart Elements): Separate function calls for title, axis labels, legend, and grid
When you run this code, a window will appear with a blue line connecting four circular points, labeled axes, a title at the top, a legend identifying the line, and a light background grid to help read values.


```python
# Import library
import matplotlib.pyplot as plt

# LAYER 1: DATA. We have two lists.
months = [1, 2, 3, 4]  # Represents the months of the year (January to April)
revenue = [100, 130, 120, 150]  # Represents the revenue for each month

# LAYER 2: OPTIONS (inside plot function) =====
plt.plot(
    months, # Months are plotted on the X-axis.
    revenue, # Revenue is plotted on the Y-axis.
    color='blue',  # Set the line color to blue.
    linestyle='-',  # Solid line style to connect the data points.
    marker='o',    # Use circular markers to highlight each data point.
    label='2024 Sales')  # Label for the legend.

# LAYER 3: CHART ELEMENTS (separate functions → titles, labels, legend, grid)
plt.title("Monthly Revenue Trend")    # Add title
plt.xlabel("Month")                   # Label X-axis
plt.ylabel("Revenue (USD)")           # Label Y-axis
plt.legend()                          # Show legend
plt.grid(True, alpha=0.3)             # Add light grid

# Show the final chart
plt.show()
```

Matplotlib is flexible and can plot data from different sources. The most common sources are: (1) Python Lists (2) NumPy Arrays (3) Pandas Series or DataFrames. All three can be used for plotting, but each has different advantages. The following script uses same data to plot from lists, Numpy Arrays and Pandas dataFrame. The following script creats plots from Python lists, Numpy Arrays and Pandas Dataframes.

```python
# Plotting from Lists, NumPy Arrays, and Pandas

import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# 1. Plotting Python Lists. Use for small data
x = [1, 2, 3, 4]  # X-axis values→ simple sequence of numbers.
y = [10, 20, 15, 30]  # Y-axis → arbitrary data points for each X
# Each call to plt.figure()→ new plot. Can create multiple visualizations..
plt.figure()  # Creates new figure (or canvas) for plotting.
plt.plot(x, y)  # Takes the X and Y data and creates a line plot.
plt.title("Python Lists")  # Add title to the plot.

plt.show()  # This function renders the plot on the screen.

# 2. Plotting NumPy Arrays. Use when maths involved
x = np.array([1, 2, 3, 4])  # List of X values→ NumPy array.
y = np.array([10, 20, 15, 30])  # List of Y values→ into a NumPy array.
plt.figure()  # Creates a new figure (or canvas) for plotting.
plt.plot(x, y)  # Takes the X and Y data and creates a line plot.
plt.title("NumPy Arrays")  # Add title to the plot.
plt.show()  # This function renders the plot on the screen.

# 3. Plotting Pandas Data. Use when data comes from a table
# Create DataFrame
df = pd.DataFrame({
    "X": [1, 2, 3, 4],
    "Y": [10, 20, 15, 30]
})
plt.figure()  # Creates a new figure (or canvas) for plotting.
plt.plot(df["X"], df["Y"]) # Take 'X' & 'Y' columns from DataFrame→ line plot
plt.title("Pandas Data")  # Add title to the plot.
plt.show()  # This function renders the plot on the screen.
```

An error bar plot displays measured values along with their uncertainties. The error bars indicate the possible range within which the actual value may lie. This helps in assessing the reliability and variability of measurements. In the example script below, a separate error value is provided for each data point

```python
import matplotlib.pyplot as plt
x = [1, 2, 3, 4]
y = [10, 20, 15, 30]
errors = [1, 2, 1, 3]  # Error values for each point in y
# So 10 is 10 ± 1, 20 is 20 ± 2, 15 is 15 ± 1, and 30 is 30 ± 3

# Create error bar plot. yerr=errors specifies the error bars for each point in y.
# fmt specifies the line and marker style.
plt.errorbar(x, y, yerr=errors, fmt="-o", color="darkorange")
plt.title("Measurement Uncertainty")
plt.show()
```

A bubble chart extends a scatter plot by varying the size of the markers. The position of each bubble is determined by the x and y values, while the s parameter controls its size. Larger bubbles represent larger values and help visualize an additional dimension of data. This is shown in the script below:

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5, 6, 7, 8]
y = [10, 20, 15, 30, 25, 45, 40, 35]
sizes = [40, 70, 100, 400, 500, 700, 800, 900]  # Bubble sizes for each point

# Create bubble chart. s=sizes specifies the size of each marker.
plt.scatter(x, y, s=sizes, color="teal", alpha=0.6)
plt.title("Bubble Sizes")
plt.show()
```

A graph may contain more than one plot at the same time. Combining different graph types often makes data easier to understand and compare.
In the following example, a bar chart is used to display sales values, while a line graph joins the tops of the bars to highlight the overall trend. The script also demonstrates how different styling options such as colors, line styles, line width, markers, and transparency can be applied to improve the appearance of a graph. Such combinations are commonly used in reports and business charts to present information more clearly.


```python
import matplotlib.pyplot as plt
x = [1, 2, 3, 4]  # X-axis→ categories or time points
sales = [10, 20, 15, 30]  # Y-axis→ sales figures for each X value
# Create bar chart
plt.bar(x, sales, color="lightblue", edgecolor="black",
        alpha=0.7, label="Sales Bars")
# Join tops of bars using line plot
plt.plot(x, sales, color="red", linestyle="--", linewidth=3, marker="o",
         markersize=8, markerfacecolor="yellow", label="Sales Trend")
plt.legend()  # Display the legend to differentiate between bars and line plot
plt.show()  # Combined bar & line chart→ shows sales data with trend line
```

The following example demonstrates how to create multiple graphs (subplots) inside the same figure using the subplots() method. It creates a 1 × 2 grid of subplots, that is: 1 row and 2 columns. The method is: fig, ax = plt.subplots(1, 2). It returns two objects: (1) fig → represents the entire figure (overall container). (2) ax → an array containing individual subplot areas
Note: Since two subplots are created, the ax variable behaves like a 1-D array with 2 items ax[0] → first subplot (left side) and ax[1] → second subplot (right side). The script then demonstrates how different graph types may be drawn in separate subplots: (1) a line graph for sales data using ax[0].plot()  (2) a bar chart for profit data using: ax[1].bar()
This is an example of the object-oriented (OOP) approach in Matplotlib, where we explicitly specify which subplot should receive the graph.
Finally, plt.tight_layout() automatically adjusts spacing between subplots so that titles and labels do not overlap or get clipped.


```python
# Import matplotlib
import matplotlib.pyplot as plt

# Create data
x = [1, 2, 3, 4]
sales = [10, 20, 15, 30]
profit = [3, 8, 5, 12]

# Create 1 row and 2 subplots. The 'fig' variable holds overall figure object
# 'ax'→ array of axes (subplots) with indices like ax[0] and ax[1]
fig, ax = plt.subplots(1, 2)

# First subplot. ax[0] refers to the first subplot (left side)
ax[0].plot(x, sales) # We plot 'sales' against 'x' using a line plot.
ax[0].set_title("Sales")

# Second subplot. ax[1] refers to the second subplot (right side)
ax[1].bar(x, profit) # We plot 'profit' against 'x' using a bar plot.
ax[1].set_title("Profit")

# Adjust subplot parameters to give specified padding
plt.tight_layout() # Prevents clipping of labels and titles.
plt.show()  # Renders the figure with both subplots displayed side by side.
```

The following example demonstrates how to improve the appearance of a graph using the object-oriented (OOP) approach of Matplotlib. The script first creates a Figure object (fig) and an Axes object (ax) using: fig, ax = plt.subplots()
The figure size and overall figure background color are specified during creation. The background color of the plotting area (subplot) is then changed separately using: ax.set_facecolor(). A line graph is created using: ax.plot(). The script does customised formatting for line color, marker style and line width
The script also demonstrates how to: (1) add a title (2) label the x-axis and y-axis (3) display gridlines and (4) improve spacing using tight_layout()
An important point to note is that the outer figure background and the inner plotting area may have different colors. This helps distinguish the overall figure from the actual graph area.


```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4]
sales = [10, 20, 15, 30]

# Create figure and subplot
fig, ax = plt.subplots(figsize=(8, 4),
    facecolor="lightyellow") # Background color of entire figure→ light yellow

ax.set_facecolor("whitesmoke") # Background color of plot area→ light smoke

# Create graph with specified line color, marker style, and line width
ax.plot(x, sales,
    color="blue", # Line color→ blue
    marker="o",   # Marker style→ circle
    linewidth=2)  # Line width→ 2 points

ax.set_title("Quarterly Sales")  # Title of the graph→ "Quarterly Sales"
ax.set_xlabel("Quarter")  # Label for x-axis→ "Quarter"
ax.set_ylabel("Sales")  # Label for y-axis→ "Sales"
ax.grid(True)  # Display gridlines
plt.tight_layout()  # Prevents clipping of labels and titles.
plt.show() # Display the graph
```

The script shows the same data in three views → The 3 plots show how different methods reveal different aspects of the same information.
Key Insight: All three panels represent the exact same numbers. The choice of visualization method depends on what you want to emphasize: exact cell values (imshow), boundaries between value ranges (contour), or smooth regional patterns
The script is as follows:-


```python
# Step 1: Import required libraries
import numpy as np
import matplotlib.pyplot as plt

# Step 2: LAYER 1 - Prepare the DATA (Self-made 5×10 peak_matrix)
# Values increase toward center (peak), decrease toward edges
peak_matrix = np.array([
    [ 5, 10, 15, 20, 25, 25, 20, 15, 10,  5], # From 5 at edges to 25 at center
    [10, 20, 35, 50, 65, 65, 50, 35, 20, 10],
    [15, 35, 60, 85,100,100, 85, 60, 35, 15],
    [10, 20, 35, 50, 65, 65, 50, 35, 20, 10],
    [ 5, 10, 15, 20, 25, 25, 20, 15, 10,  5]])

# Step 3: Create figure with 1 row and 3 columns
fig, axs = plt.subplots(1, 3, figsize=(12, 4))

# Step 4: LAYER 2 - Subplot 1: imshow() (Colored Cells)
im = axs[0].imshow(peak_matrix,
    cmap='viridis',  # Color scheme: low=dark, high=bright
    interpolation='nearest')  # Sharp cell boundaries
axs[0].set_title("imshow(): Colored Cells")
axs[0].set_xlabel("Column")
axs[0].set_ylabel("Row")
plt.colorbar(im, ax=axs[0])

# Step 5: LAYER 2 - Subplot 2: contour() (Lines Only)
contour = axs[1].contour(peak_matrix,
    levels=6,              # 6 contour lines for clear rings
    colors='black',        # Line color
    linewidths=1.5)         # Line thickness
axs[1].set_title("contour(): Lines Only")
axs[1].set_xlabel("Column")
axs[1].set_ylabel("Row")
axs[1].grid(True, alpha=0.3)

# Step 6: LAYER 2 - Subplot 3: contourf() (Filled Regions)
contour_filled = axs[2].contourf(peak_matrix,
    levels=6,              # 6 filled regions for clear rings
    cmap='viridis')         # Color scheme for fills
axs[2].set_title("contourf(): Filled Regions")
axs[2].set_xlabel("Column")
axs[2].set_ylabel("Row")
plt.colorbar(contour_filled, ax=axs[2])

# Step 7: LAYER 3 - Add overall chart elements
fig.suptitle("Hand-Crafted Peak Matrix: Three Views",
             fontsize=14, fontweight='bold', y=1.02)

# Step 8: Adjust layout and display
plt.tight_layout()
plt.show()

# Step 9: Print matrix for reference (compact format)
print("Peak Matrix (5 rows × 10 columns):")
for row in peak_matrix:  # Print each row of the matrix
    # " ".join(...) creates a string of values separated by spaces
    # Format each value to be 3 characters wide for alignment
    print(" ".join(f"{v:3d}" for v in row)) # .
```

The following example combines three useful Matplotlib features.
•	First, a built-in style (ggplot) is applied to improve the overall appearance of the graph.
•	Next, a text annotation is added to highlight an important data value.
•	Finally, the graph is saved as an image file using savefig() before being displayed on the screen.
The example also prints all available styles on terminal using print(plt.style.available). It further demonstrates that styles are applied globally through plt.style.use(), while graph creation and annotation are performed using the object-oriented (fig, ax) approach.


```python
## Combined Example: Styles, Text Annotations and Saving Graphs
import matplotlib.pyplot as plt

# Print available styles on terminal & Apply a built-in style (theme)
print(plt.style.available)  # List of available styles→ 'ggplot', 'seaborn'...
plt.style.use("ggplot")  # ggplot style mimics the aesthetics of R's ggplot2

products = ["A", "B", "C", "D"]  # Product categories→ (Nominal Scale)
sales = [25, 40, 20, 15]  # Sales figures for each product (Ratio Scale)

fig, ax = plt.subplots(figsize=(6, 4)) # Single plot canvas (6 wide X 4 tall)

# Create bar chart with custom colors and transparency
ax.bar(products, sales,color="yellow", edgecolor="black", alpha=0.7)

# Add text annotation to highlight highest sale
ax.text(
    1,                      # x-coordinate (Product B)
    40,                     # y-coordinate (Sales value)
    "Highest Sale",         # Text to display
    fontsize=10,
    color="red",
    ha="center"   # Horizontal alignment: center the text above the bar
)

ax.set_title("Product Sales")  # Graph title
ax.set_xlabel("Product (Nominal Scale X Axis)→") # X-axis label & scale type
ax.set_ylabel("Sales (Ratio Scale Y Axis)→")  # Y-axis label & scale type
# Save graph as PNG image. File extension determines format.
fig.savefig("sales_chart.png", dpi=300) # Save at 300 dots-per-inch.
plt.show()  # Display the graph on screen
```


The following example demonstrates how a histogram and a box plot can be used together to understand the distribution of a dataset. The script first generates approximately bell-shaped (normally distributed) exam scores using NumPy. A few extreme values are then added intentionally so that the effect of outliers can be observed in both the histogram and the box plot. The script calculates the three quartiles: (1) Q1 (25th percentile). (2) Q2 (Median or 50th percentile) and (3) Q3 (75th percentile). It then displays them on the histogram using vertical reference lines. These quartiles divide the data into four equal parts and form the basis of the box plot. To make comparison easier, the figure contains two subplots:
•	a histogram in the upper portion showing the detailed frequency distribution
•	a box plot in the lower portion showing a compact statistical summary
The parameter: gridspec_kw={"height_ratios":[3,1]} allocates more vertical space to the histogram than to the box plot, since the histogram requires greater display area.
An important observation is that both graphs represent the same dataset:
•	the histogram shows the overall shape of the distribution
•	the box plot summarizes the distribution using quartiles, median and outliers
Thus, the example helps illustrate the relationship between:
Data → Histogram → Quartiles → Box Plot
It shows how box plots provide a compact summary of information that is also visible in a histogram.


```python
import numpy as np
import matplotlib.pyplot as plt

# Generate approx bell shaped data & calculate Q1, Q2 (Median), Q3
np.random.seed(10)  # Set seed for reproducibility
# loc=70→ average/ mean score set to 70, scale=10. Standard Deviation→ 10
scores = np.random.normal(loc=70, scale=10, size=100)
# Add 3 extreme values (outliers)→ to show affect on box plot and histogram
scores = np.append(scores, [25, 30, 105])
# np.percentile() computes the q-th percentile of the data.
q1 = np.percentile(scores, 25)  # 25th percentile (Q1)
q2 = np.percentile(scores, 50)  # 50th percentile (Median)
q3 = np.percentile(scores, 75)  # 75th percentile (Q3)

# Create figure with two subplots: histogram on top, box plot on bottom
# gridspec_kw={"height_ratios":[3,1]}→ histogram 3 parts, box plot 1 part
fig, ax = plt.subplots(2, 1, figsize=(8, 6),
                       gridspec_kw={"height_ratios":[3,1]})

# Histogram. bins=10→ divide score range into 10 intervals.
ax[0].hist(scores, bins=10, color="skyblue", edgecolor="black")

# Quartile lines, title, labels, and grid for histogram
ax[0].axvline(q1, color="red", linestyle="--", label="Q1")
ax[0].axvline(q2,color="green",linestyle="-",label="Median")
ax[0].axvline(q3,color="red",linestyle="--",label="Q3")
ax[0].set_title("Histogram and Box Plot of Exam Scores")
ax[0].set_ylabel("Frequency")
ax[0].legend()
ax[0].grid(axis="y", alpha=0.3)

# Box Plot. vert=False→ horizontal box plot, showmeans=True→ show mean as
# a point, patch_artist=True→ allow custom colors
ax[1].boxplot(scores,vert=False,showmeans=True,patch_artist=True)
ax[1].set_xlabel("Marks")
ax[1].set_yticks([])  # Remove y-axis ticks for cleaner look

# Display Quartile Values
print("Q1 =", round(q1, 2))
print("Median =", round(q2, 2))
print("Q3 =", round(q3, 2))

plt.tight_layout()  # To prevent clipping of labels and titles
plt.show()  # Render the combined histogram and box plot visualization
```
