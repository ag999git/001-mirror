What Does re.escape() Do? This can be best understood by an example. Suppose a user enters the following text: a+b. You want to match this text literally. However, in Regular Expressions: +
is a special metacharacter, meaning: “one or more occurrences”. Therefore: re.fullmatch("a+b", "a+b") fails. The failure is because Python interprets: "a+b" as a regex pattern, not ordinary text.
The regex engine reads it as: “Match one or more a characters followed by b.”
So: a+b is not treated literally.
The Solution: Use re.escape(). We can convert special regex characters into ordinary literal characters. Example:


```python
import re
safe_pattern = re.escape("a+b")
print(safe_pattern)  # Output: a\+b
# re.escape() function takes string "a+b" & escapes the special character "+"
# by adding a backslash before it, resulting in the escaped string "a\+b".
print(re.fullmatch(safe_pattern, "a+b")) # Success
# Output: <re.Match object; span=(0, 3), match='a+b'>
```

A Useful Rule of Thumb: When designing regex groups: (1) Place ( ) only around the data you want to extract. (2) Leave navigation text (spaces, separators, punctuation, filler text, etc.) outside the brackets if you do not want to store it.
So if we want to capture parts of search filter, we do the following:-
•	Use () only for the specific data you want to extract.
•	Leave parts strictly for navigation (like spaces, separators, or filler text) outside the brackets.
This is best understood by an example script:


```python
import re
text = "abc 123 def 456"

# Example 1: Capture ONLY the two digit sequences. Ignore text between them.
# The pattern is r"(\d+)\D+(\d+)".
# (\d+) captures first sequence of digits (the first group),
# \D+ matches space & "def" between "123" & "456". Does not capture them as group.
# (\d+) matches and captures the second sequence of digits (456) (the second group).

m1 = re.search(r"(\d+)\D+(\d+)", text)
print("First digits :", m1.group(1))   # 123
print("Second digits:", m1.group(2))   # 456

# Example 2: Capture digits + intervening text + digits
# The pattern is r"(\d+)(\D+)(\d+)".
# (\d+) captures first sequence of digits (123) (group 1)
# (\D+) captures space & "def" between "123" & "456" (group 2)
# (\d+) captures second sequence of digits (456) (group 3)

m2 = re.search(r"(\d+)(\D+)(\d+)", text)
print("First digits :", m2.group(1))   # 123
print("Non-digits   :", m2.group(2))   # " def "
print("Second digits:", m2.group(3))   # 456
```

11.14.	Script using convenience functions
The script given below shows the use of 8 convenience functions of the re module.
The purpose and usage of each of these 8 functions is explained in comments. Moreover, after every print() function there is a comment (#) below it which shows the expected output.


```python
import re

# Sample text used for all examples
text = "abc 123 def 456"

# 1. re.match()
# Tries to match the pattern ONLY at the beginning of the string
result = re.match(r"abc", text)  # Match 'abc' at start
print(f"Match {r'abc'} With {text}-> {result}")
# Match abc With abc 123 def 456-> <re.Match object; span=(0, 3), match='abc'>

# 2. re.fullmatch()
# Matches the pattern against the ENTIRE string
result = re.fullmatch(r"abc 123 def 456", text) #Fullmatch entire string
print(f"Fullmatch {r'abc 123 def 456'} With {text}-> {result}")
# Fullmatch abc 123 def 456 With abc 123 def 456-> <re.Match object; span=(0, 15),           match='abc 123 def 456'>

# 3. re.search()
# Searches for the pattern ANYWHERE in the string
result = re.search(r"123", text)  # Search for '123'
print(f"Search {r'123'} In {text}-> {result}")
# Search 123 In abc 123 def 456-> <re.Match object; span=(4, 7), match='123'>

# 4. re.findall()
# Finds ALL matches and returns them as a LIST
# r"\d+" matches sequences of digits
result = re.findall(r"\d+", text)  # Find all digit sequences
print(f"Findall {r'\d+'} In {text}-> {result}")
# Findall \d+ In abc 123 def 456-> ['123', '456']

# 5. re.finditer()
# Finds ALL matches and returns an ITERATOR
# Each item is a Match object
print("re.finditer(r'\d+', text):")
for match in re.finditer(r"\d+", text):  # Iterate over all matches
    print("Match:", match.group(), "at position", match.start())
# Match: 123 at position 4
# Match: 456 at position 12

# 6. re.split()
# Splits the string wherever the pattern matches
# r"\s+" matches one or more whitespace characters
result = re.split(r"\s+", text)  # Split on whitespace
print(f"Split {r'\s+'} In {text}-> {result}")
# Split \s+ In abc 123 def 456-> ['abc', '123', 'def', '456']

# 7. re.sub()
# Replaces ALL matches with a new value
result = re.sub(r"\d+", "#", text)  # Replace digit sequences with '#'
print(f"Sub {r'\d+'} With '#' In {text}-> {result}")
# Sub \d+ With '#' In abc 123 def 456-> abc # def #

# 8. re.subn()
# Same as sub(), but ALSO returns the count of replacements
result = re.subn(r"\d+", "#", text)  # Replace digit sequences with '#' and get count
print(f"Subn {r'\d+'} With '#' In {text}-> ", result)
# Subn \d+ With '#' In abc 123 def 456->  ('abc # def #', 2)
```

11.15.	Script Using Instance Methods of the Pattern Object
We discussed earlier that the Pattern object provides several instance methods for matching, searching, splitting, and substitution. To use these methods, we must first create a Pattern object using the factory function: re.compile()
Unlike convenience functions such as re.search(), instance methods operate on an already compiled regex pattern, making them reusable and potentially more efficient.
The following script demonstrates some commonly used instance methods of the Pattern object.



```python
import re

# Sample text used throughout the script
text = "abc 123 def 456"

# Create 3 Pattern objects (compile once, reuse later)
p_digits = re.compile(r"\d+")   # One or more digits
p_start  = re.compile(r"abc")   # Match at start
p_full   = re.compile(r"abc 123 def 456")

# 1. match(). Checks ONLY at the beginning of the string
result = p_start.match(text)
print("1. match() ->", result)
# OUTPUT: <re.Match object; span=(0, 3), match='abc'>

# 2. fullmatch(). Entire string must match the pattern
# If entire string matches, returns a Match object; otherwise, returns None.
result = p_full.fullmatch(text)
print("2. fullmatch() ->", result)
# OUTPUT: <re.Match object; span=(0, 15), match='abc 123 def 456'>

# 3. search(). Searches for pattern anywhere in the string
# If pattern is found, returns a Match object; otherwise, returns None.
result = p_digits.search(text)
print("3. search() ->", result)
# OUTPUT: <re.Match object; span=(4, 7), match='123'>

# 4. findall(). Returns ALL matches as a list

result = p_digits.findall(text)
print("4. findall() ->", result)
# OUTPUT: # ['123', '456']

# 5. finditer(). Returns an iterator yielding Match objects for all matches
print("5. finditer() ->")
for m in p_digits.finditer(text):  # Return Match objects one by one
    print(m.group(), "found at", m.span())
# OUTPUT:
# 123 found at (4, 7)
# 456 found at (12, 15)

# 6. split(). The pattern \d+ matches digit sequences and split() breaks the
# string wherever these digits occur. The matched digits are removed from the
# result, while non-matching text (including spaces) is preserved. The final
# empty string appears because splitting also occurs after the last match ("456")
result = p_digits.split(text)
print("6. split() ->", result)
# OUTPUT: ['abc ', ' def ', '']  # Note that a final empty string appears

# 7. sub(). Replaces matches with new text
result = p_digits.sub("$", text)
print("7. sub() ->", result)
# OUTPUT: abc $ def $

# 8. subn(). # Same as sub(), but also returns replacement count
result = p_digits.subn("$", text)
print("8. subn() ->", result)
# OUTPUT: ('abc $ def $', 2)  # Count of replacements is 2

```

The following script shows use of if/ else block or try/ except block while working with Match objects. In this script, the Match object namely m_match is None, so we cannot use its attributes. We should always test the Match object against None, or else wrap the code in try/ except block to prevent potential errors.

```python
import re
m_match = re.match(r"(\d+)", "abc 123 def 456")
# m_match will be None since there's no match at the start of the string

# Method 1. Using an if-statement to check for None
if m_match is None:
    print("No match found for re.match.")  # This will be executed
else:
    print(m_match.group(1))  # This won't be executed

# Method 2. Using try-except to handle potential AttributeError
try:
    print(m_match.group(1))  # Access group even though m_match is None
except AttributeError as e:
    print("Caught an AttributeError:", e)  # Handle m_match is None
    print(type(e))  # Here it will print <class 'AttributeError'>

```

The instance methods of a Pattern object perform operations such as: (1) pattern.match() (2) pattern.search() and (3) pattern.findall(). But the attributes of a Pattern object describe the compiled regex itself. For example:

```python
import re
pattern = re.compile(r"(?P<word>\w+)-(\d+)")
print(pattern.pattern)  # (?P<word>\w+)-(\d+)
print(pattern.groups)   # 2
print(pattern.groupindex)  # {'word': 1}
print(pattern.flags)    # 32
```

The following script shows how some of the common attributes of the Pattern object are used:

```python
# 1. pattern. Returns original regex text from which Pattern object was created.
import re
pattern = re.compile(r"\d+")

print(pattern.pattern)  # \d+. Gets the original regex source.

# 2. groups. Returns the total number of capturing groups.
pattern = re.compile(r"(\d+)-(\w+)")

print(pattern.groups)  # 2. because there are two capturing groups.

# 3. groupindex. Returns a dictionary mapping group name → group number.
pattern = re.compile(r"(?P<word>\w+)-(\d+)")
print(pattern.groupindex)  # {'word': 1}. Meaning: word → group 1

# 4. flags. Returns the regex compilation flags.
pattern = re.compile(r"abc", re.I)
print(pattern.flags)  # 34. Includes re.IGNORECASE internally.
```

The following script shows the difference between greedy and non-greedy regex:-



```python
import re
text = "abc123xyz456"
# 1. Greedy matching
m1 = re.search(".*\\d+", text) # greedy
print(m1.group())  # Output: abc123xyz456

# 2. Non-greedy (lazy) matching by adding ? after the quantifier *
# ".*?\\d+" is non greedy because .*? matches as little text as possible
m2 = re.search(".*?\\d+", text) # non-greedy
print(m2.group())  # Output: abc123

# Given below is another example. In the given string namely str_html, the first closing angular bracket  > appears at the end of <!DOCTYPE html>. So non-greedy (lazy) will stop here.
# On the other hand, the last angular bracket > appears at </html>, so greedy will go up to that last angular bracket.

import re
str_html = '<!DOCTYPE html> <html> <head> </head> </html>'
print('length string->', len(str_html))  # length of the string-> 45

# 1.GREEDY.Does not have ? after *. This makes it greedy (default)
pat_greedy = '<.*>' #
print('length greedy->', re.match(pat_greedy, str_html).span())
# Output: length greedy-> (0, 45)
print('groups greedy->', re.match(pat_greedy, str_html).group())
# Output: <!DOCTYPE html> <html> <head> </head> </html>

# 2. NON-GREEDY. It has ? after * which makes it non-greedy (lazy)
pat_nongreedy = '<.*?>' #
print('length non-greedy->', re.match(pat_nongreedy, str_html).span())
# Output: length non-greedy-> (0, 15)
print('groups non-greedy->', re.match(pat_nongreedy, str_html).group())
# Output: <!DOCTYPE html>
```

Why Use Non-Capturing Groups? Non-capturing groups are useful when: (1) grouping is needed. (2) captured text is not needed
They help: (1) keep group numbering simple. (2) make regex easier to read. (3) avoid unnecessary captures. The following examples will clarify the concepts:


```python
import re
# 1.    Example 1: Grouping for Repetition
text = "ababab"
# (?:ab)+ is non-capturing group (?:...) it doesnt create numbered group for
# the match, but allows application of + quantifier to the whole "ab" sequence
print(result)  # Output: ['ababab']
result = re.findall(r"(?:ab)+", text)

# 2.    Example 2: Grouping for Alternation
text = "cat dog rat bat"
# (?:cat|dog|rat) is non capturing group (?:...) so doesn't create numbered
# group for the match. The | operator matches any of cat, dog, or rat.
result = re.findall(r"(?:cat|dog|rat)", text)
print(result)  # Output: ['cat', 'dog', 'rat']
# Explanation: The group is used only to combine alternatives.

# 3.    Example 3: Cleaner Group Numbering
# Explaining: (?:Mr|Ms|Dr)\. (\w+)
# (?:Mr|Ms|Dr)→ non-capturing group→ matches title but doesnt create numbered group
# while (\w+) captures the name as Group 1.
m = re.search(r"(?:Mr|Ms|Dr)\. (\w+)", "Dr. John")
print(m.group(1))  # Output: John

# 4.    Example 4: Removing Duplicate Words
text = "The The cat cat sat sat"
# (\b\w+)(?:\s+\1)+ → (\b\w+) captures a word as Group 1, and (?:\s+\1)+ matches
# one or more occurrences of same word preceded by whitespace, but doesnt capture
# them as a group. The re.sub() function replaces the entire matched sequence
# with just the first occurrence of the word (Group 1)→ removing duplicates
pattern = r"(\b\w+)(?:\s+\1)+"
result = re.sub(pattern, r"\1", text)
print(result)  # Output: The cat sat
```

What Are Backreferences?:- When parentheses () are used in a regex, then \1 refers to the first group, \2 refers to the second group and so on. Backreferences allow us to reuse matched text during replacement.
The following script shows how the FirstName LastName can be reordered to LastName FirstName.


```python
import re  # Import the regular expression module

# Pattern explanation:
# ([A-Za-z]+)  → Group 1: First name (only letters)
# \s+          → One or more spaces
# ([A-Za-z]+)  → Group 2: Last name
name_pattern = r'([A-Za-z]+)\s+([A-Za-z]+)'

# Sample text containing multiple names
text = "Doe, John; Smith, Alice"

# Replace using backreferences:
# \2 → Last name
# \1 → First name
# Output format: LastName, FirstName
result = re.sub(name_pattern, r'\2, \1', text)

# Display the result
print(result)
# Output: John, Doe; Alice, Smith
```

11.25.	Removing Consecutive Duplicate (or more occurrences of) Words
The following script shows how multiple occurrences of a word can be removed using regex and backreferences.


```python
import re
# Import Python's regular expression module

# Sample text with consecutive duplicate words
text = "cat cat cat dog dog fish fish fish fish bird"

# Regex pattern to match consecutive duplicate words
pattern = r'(\b\w+)(?:\s+\1)+'
# Pattern explanation:
# (\b\w+)      → Captures a complete word and stores it as Group 1
#     \b       → Word boundary (ensures full word match)
#     \w+      → One or more word characters (letters, digits, underscore)
# (?: )        → Non-capturing group (used only for grouping, not storing)
#     \s+      → One or more whitespace characters
#     \1       → Backreference to the word captured in Group 1
#     +        → Matches one or more repetitions of the same word
#
# Overall meaning:
# "Match a word followed by one or more repetitions of the same word"

# Replace duplicate word sequences with a single occurrence
cleaned_text = re.sub(pattern, r'\1', text)
# re.sub() searches for the pattern and replaces it
# r'\1' means we keep only the first captured word
# All repeated copies are removed

print("Cleaned text:-> ", cleaned_text)
# Output: Cleaned text:->  cat dog fish bird
```

11.26.1. Using lookaround to ask user to select a valid password
The following script shows use of look around for validating a password. Rules for a valid password are:- (1) At least 8 characters. (2) At least one uppercase letter. (3) At least one lowercase letter


```python
import re
# Regex with lookarounds
# ^ asserts position at start of the string
# .* means any character (.) zero or more times (*)
# Password rules using lookahead assertions:
# (?=.*[A-Z]) -> Any character 0 or more times. Then 1 uppercase letter
# (?=.*[a-z]) -> Any character 0 or more times. Then 1 lowercase letter
# .{8,}       -> at least 8 characters total

password_pattern = re.compile(r"^(?=.*[A-Z])(?=.*[a-z]).{8,}$")

# Test passwords
passwords = [
    "no_uppercase",     # no uppercase
    "NO_LOWERCASE",     # no lowercase
    "Short",            # too short
    "Password",         # valid
    "GoodPass1"         # valid
]

# Test each password
for pwd in passwords:  # Check each password
    if password_pattern.search(pwd):  # Valid password
        print(pwd, "-> VALID")
    else:  # Invalid password
        print(pwd, "-> INVALID")
```

Is it possible to write this password validation without using lookaround? Yes. But it becomes very complicated. The pattern with no lookaround will be
pattern_no_lookaround = re.compile( r"(^[A-Z].*[a-z].{6,}$)|(^[a-z].*[A-Z].{6,}$)")
The script which implements above password check without using lookaround is as follows:-


```python
import re

# REGEX WITHOUT LOOKAROUNDS
# Explanation:
# 1. Minimum length 8 → .{8,}
# 2. Must contain:
#    a) uppercase before lowercase OR
#    b) lowercase before uppercase
# This forces us to write BOTH cases using alternation (|)
# Upper case before lowercase: ^[A-Z].*[a-z].{6,}$
# Lower case before uppercase: ^[a-z].*[A-Z].{6,}$

condition1 = "^[A-Z].*[a-z].{6,}$"
condition2 = "^[a-z].*[A-Z].{6,}$"
pattern_no_lookaround = re.compile(rf"({condition1})|({condition2})")

# Passwords to test
passwords = [
    "no_uppercase",     # no uppercase
    "NO_LOWERCASE",     # no lowercase
    "Short",            # too short
    "Password",         # valid
    "GoodPass1"         # valid
]

# Test each password
for pwd in passwords:
    if pattern_no_lookaround.search(pwd):
        print(pwd, "-> VALID")
    else:
        print(pwd, "-> INVALID")

# Output:
# no_uppercase  -> INVALID
# NO_LOWERCASE  -> INVALID
# Short         -> INVALID
# Password      -> VALID
# GoodPass1     -> VALID
```

Suppose you have to search for sub-string "ana" in the string "banana". There are actually two "ana" sequences:
1.	b ana na (starting at index 1)
2.	ban ana (starting at index 3)
The problem with standard matching:- If you use re.findall(r"ana", "banana"), the engine finds the first ana, and the cursor moves to index 4. It has "consumed" the middle a, so it can't use it to start the second ana.
The Solution: The Lookahead Capture. To find overlaps, we use a Positive Lookahead containing a Capturing Group. The lookahead checks the condition (is there an "ana" here?) without moving the cursor, and the parentheses "snap a photo" of what it saw.


```python
import re

text = "banana"
# We look for a position that is followed by 'ana'
# The lookahead (?=...) has zero width, so the cursor stays put!
pattern = r"(?=(ana))"  # Lookahead to find 'ana' without consuming characters

matches = re.finditer(pattern, text)  # Find all matches in the text
# We use re.finditer to get an iterator over all matches
# matches will find overlapping occurrences of 'ana' in 'banana'
# Each match has zero width, but captures 'ana' in group 1
for m in matches:
    # Match object for each occurrence
    print("Match object:-> ", m)
    # m.group(0) is empty because the lookahead consumed 0 characters!
    # m.group(1) contains the 'ana' captured inside the lookahead.
    content = m.group(1)  # The matched content
    index = m.start()  # The starting index of the match
    # span wont be accurate since the lookahead has zero width
    print("span->", m.span()) #
    print(f"Found '{content}' starting at index {index}")
    print('-' * 40) # Separator for clarity

# Output:
# Match object:->  <re.Match object; span=(0, 0), match=''>
# span-> (0, 0)
# Found 'ana' starting at index 0
# Match object:->  <re.Match object; span=(1, 1), match=''>
# span-> (1, 1)
```

There are real situations where: Lookahead cannot solve the problem at all, but lookbehind solves it cleanly. Example: Match something ONLY if it is preceded by a pattern — but do NOT include that pattern. This is the classic case where: Lookbehind works but Lookahead cannot work
For example suppose we have a string:- "Ms. Alice and Mr. Bob met Dr. Charlie."
Constraint
•	Titles (Mr., Ms., Dr.) must not appear in the match
•	The name must be matched only if it is preceded by a title
So we should have names like [Alice, Bob, Charlie]
This problem cannot be solved by lookahead. Lookahead checks what comes AFTER the current position. The following script shows how we can do the above task


```python
import re

text = "Ms. Alice and Mr. Bob met Dr. Charlie."

# (?<=...) is a POSITIVE LOOKBEHIND
# It asserts that before the current position we have:
#   Mr.  OR  Ms.  OR  Dr.
# Important: The title is CHECKED but NOT INCLUDED in the match
# sp stands for space character. Just a single space here.
# - sp is represented by \s in regex
# - Only the name is captured

pattern = r"(?<=(Mr|Ms|Dr)\.\s)[A-Z][a-z]+"
matches = re.findall(pattern, text)
print(matches) # OUTPUT: ['Alice', 'Bob', 'Charlie']

```

It is possible to write a regular expression which checks for prime number. The script which does so is given below (Using backreference):-



```python
import re
def is_prime(n):
    pat_for_prime = r'^1?$|^(11+?)\1+$'
    mo = re.match(pat_for_prime, "1" * n)
    if mo == None:
        return True
    else:
        return False
print('is prime 7->', is_prime(7))  # True
print('is prime 9->', is_prime(9))  # False
print('is prime 11->', is_prime(11))  # True
```
